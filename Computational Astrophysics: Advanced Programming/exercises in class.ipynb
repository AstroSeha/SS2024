```python
from astropy.io import fits
import numpy as np
import matplotlib.pyplot as plt
```

```python
r_file = '/Users/sehalee/Downloads/frame-r-006182-2-0256.fits'
r_frame = fits.getdata(r_file)
print(r_frame.shape)
```

```python
plt.imshow(r_frame, cmap='gray',norm = 'log', vmin = 1e-2, vmax = 1e2)
plt.colorbar()
plt.show()
```

```python
#crop the image
r_frame_cropped = r_frame[450:1050, 1000:1800]
plt.imshow(r_frame_cropped, cmap='gray',norm = 'log', vmin = 1e-2, vmax = 1e2)
```

```python
# do the same for u,g,r,i,z

frame = np.zeros((5, 600, 800))
frame[0] = fits.getdata('/Users/sehalee/Downloads/frame-u-006182-2-0256.fits')[450:1050, 1000:1800]
frame[1] = fits.getdata('/Users/sehalee/Downloads/frame-g-006182-2-0256.fits')[450:1050, 1000:1800]
frame[2] = fits.getdata('/Users/sehalee/Downloads/frame-r-006182-2-0256.fits')[450:1050, 1000:1800]
frame[3] = fits.getdata('/Users/sehalee/Downloads/frame-i-006182-2-0256.fits')[450:1050, 1000:1800]
frame[4] = fits.getdata('/Users/sehalee/Downloads/frame-z-006182-2-0256.fits')[450:1050, 1000:1800]
```

```python
# 4 plots except for r

fig, axs = plt.subplots(2, 2)

for i in range(4):
    axs[i//2, i%2].imshow(frame[i], cmap='gray', norm='log', vmin=1e-2, vmax=1e2)
    axs[i//2, i%2].axis('off')
    axs[i//2, i%2].set_title('ugiz'[i])

plt.tight_layout()
plt.show()
```

```python
value = np.zeros((5,2))
```

```python
# u
i = 0
a= 1e-2
b= 0.4
value[i,0] = a 
value[i,1] = b

fig, axs = plt.subplots(1, 2)

axs[0].hist(frame[i].flatten(), range=(0,1))
axs[1].imshow(frame[i], cmap='gray', norm='log', vmin = a, vmax = b)

plt.show()
```

```python
# g
i = 1
a= 1e-2
b= 0.75
value[i,0] = a 
value[i,1] = b

fig, axs = plt.subplots(1, 2)

axs[0].hist(frame[i].flatten(), range=(0,1))
axs[1].imshow(frame[i], cmap='gray', norm='log', vmin = a, vmax = b)

plt.show()
```

```python
# r
i = 2
a= 1e-2
b= 1.1
value[i,0] = a 
value[i,1] = b

fig, axs = plt.subplots(1, 2)

axs[0].hist(frame[i].flatten(), range=(0,1))
axs[1].imshow(frame[i], cmap='gray', norm='log', vmin = a, vmax = b)

plt.show()
```

```python
# 3*2 subplots of 5 frames

fig, axs = plt.subplots(3, 2, figsize=(10, 15))

for i in range(5):
    axs[i//2, i%2].imshow(frame[i], cmap='gray', norm='log', vmin=value[i,0], vmax=value[i,1])
    axs[i//2, i%2].axis('off')
    axs[i//2, i%2].set_title('ugriz'[i]+': range [{},{}]'.format(value[i,0], value[i,1]))

plt.legend()
plt.tight_layout()
plt.show()
```
