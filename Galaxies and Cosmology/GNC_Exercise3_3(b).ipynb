```python
import numpy as np
import matplotlib.pyplot as plt

# Constants
_x = 1
_y = np.pi / 2

t = np.arange(-10, 10.1, 0.1)  
A_x = 2 
A_y = 1 


x_t = A_x * np.cos(_x * t)
y_t = A_y * np.cos(_y * t)

vx_t = -A_x * _x * np.sin(_x * t)
vy_t = -A_y * _y * np.sin(_y * t )

Lz_t = x_t * vy_t - y_t * vx_t

# The moment when Lz changes sign
Lz_0 = []
for i in range(len(t) - 1):
    if Lz_t[i] * Lz_t[i + 1] < 0:
        if abs(Lz_t[i]) < abs(Lz_t[i + 1]):
            Lz_0.append(i)
        else:
            Lz_0.append(i + 1)

# Plots
plt.figure(figsize=(12, 5))
ax1 = plt.subplot(121)
plt.scatter(x_t, y_t, label='Position', color='#008B8B', s = 20)

for i in Lz_0:
    plt.scatter(x_t[i],y_t[i], color='#FF6347', s=50)

for i in range(2):
    plt.text(x_t[Lz_0[i]], y_t[Lz_0[i]], f't={t[Lz_0[i]]:.2}', horizontalalignment='right',fontsize='large')
plt.text(x_t[Lz_0[2]], y_t[Lz_0[2]], f't={t[Lz_0[2]]:.2}', horizontalalignment='left',fontsize='large')

plt.xlabel('x(t)',fontsize='large')
plt.ylabel('y(t)',fontsize='large')
plt.title('Position', fontsize='x-large')

ax2 = plt.subplot(122)
plt.scatter(t, Lz_t, label='Angular momentum', color='#6A5ACD', s = 20)
for i in Lz_0:
    plt.scatter(t[i],Lz_t[i], color='#FF6347', s=50)
    plt.text(t[i], Lz_t[i], f't={t[i]:.2}', horizontalalignment='right',fontsize='large')

plt.xlabel('t',fontsize='large')
plt.ylabel('Lz(t)',fontsize='large')
plt.title('Angular momentum', fontsize='x-large')
plt.axhline(y=0, color='#D5D8DC')
plt.tight_layout()
```
